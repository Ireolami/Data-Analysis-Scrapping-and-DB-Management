```python
import pandas as pd
import numpy as np
```

Reading of Dataset

```python
df = pd.read_csv("Dataset/Melbourne_housing_FULL.csv")
```

```python
df
```

```python
#Setting Suburb as the index
df.set_index("Suburb", inplace =True)
df
```

```python
#Dropping unnecessary data
new_df =df.drop(columns =["Type", "Method", "Postcode", "Lattitude", "Longtitude", "Regionname"])
new_df
```

```python
#Extracting the data where price is notnull
new_data = new_df[new_df["Price"].notnull()]
```

```python
new_data
```

```python
#fill missing numbers in bedroom2 using number of rooms
new_data["Bedroom2"].fillna(new_data["Rooms"], inplace= True)
new_data
```

```python
#filling missing numbers in Bathroom using Bedroom2
new_data["Bathroom"].fillna(new_data["Bedroom2"], inplace=True)
```

```python
new_data
```

```python
#filling the missing number of values of the building area with the mean
mean_ba = new_data["BuildingArea"].mean()
```

```python
new_data["BuildingArea"].fillna(mean_ba, inplace=True)
new_data
```

```python
#filling missing numbers in year using the mean values
mean_yb = new_data["YearBuilt"].mean()
mean_yb
```

```python
new_data["YearBuilt"].fillna(mean_yb, inplace=True)
new_data
```

```python
#Trying to get all data where the landsize is not empty
full_land= new_data[new_data["Landsize"].notnull()]
full_land
```

```python
#Extracting price, bedroom2 and landsize
bed_landsize_price = new_data[["Price", "Bedroom2", "Landsize"]]
```

```python
bed_landsize_price 
```

```python
#Finding the relationship between bedroom, landsize and price
relationship_b_l_and_price =bed_landsize_price.corr(method="pearson")
```

```python
relationship_b_l_and_price
```

#The relationship between the number of bedroom and Price is a positive relationship, although not strong. This 
#indicate that as the bedroom increases, the price increases, while the relationship between landsize and price 
#is none as the value obtain (0.032) is a value that can be approximated to zero

```python
#finding the relationship between year built, building area and price
yb_ba_price = new_data[["Price", "BuildingArea", "YearBuilt"]]
yb_ba_price
```

```python
correlating = yb_ba_price.corr(method="pearson")
correlating
```

```python
#extracting the price and bedroom2
bed_price = new_data[["Price", "Bedroom2"]]
bed_price
```

```python
#grouping by bedroom2 and finding the sum
sum_= bed_price.groupby("Bedroom2").sum()
sum_
```

```python
#grouping by bedroom2 and finding the mean
mean_=bed_price.groupby("Bedroom2").median()
mean_
```

```python
#Merging the mean and the sum in a dataframe
mean_sum =pd.merge(sum_, mean_, left_index=True, right_index=True)
mean_sum
```

```python
#Renaming the table
mean_sum.columns=["Sum", "Median"]
```
