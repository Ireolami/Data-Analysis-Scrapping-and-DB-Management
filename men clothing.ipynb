```python
from bs4 import BeautifulSoup as soup
import numpy as np
```

```python
from urllib.request import urlopen as uReg
```

```python
my_url ="https://www.flipkart.com/clothing-and-accessories/pr?sid=clo&otracker=categorytree&p%5B%5D=facets.ideal_for%255B%255D%3DMen&otracker=nmenu_sub_Men_0_Clothing"
uClient=uReg(my_url)
page_html=uClient.read()
uClient.close()
```

```python
page_soup=soup(page_html, "html.parser")
page_soup
```

```python
containers=page_soup.findAll("div", {'id': "container"})
```

```python
prettified=containers[0].prettify()
```

```python
prettified
```

```python
# container =contents[0]
contents= containers[0].findAll('div', {'class': '_13oc-S'})
len(contents)
```

```python
container = contents[0]
container
```

```python
products_names = []
product_price = []
product_discount = []
for i in contents:
    product=i.findAll('a', {'class': "IRpwTa"})
    if len(product) == 0:
        products_names.append(np.nan)
    else:
        name=product[0].text
        products_names.append(name)
    price = i.findAll("div", {'class':'_30jeq3'}) 
    if len(price)==0:
        product_price.append(np.nan)
    else:
        pri= price[0].text
        product_price.append(pri)
    discount=i.findAll('div', {'class': '_3Ay6Sb'})
    if len(discount)==0:
        product_discount.append(np.nan)
    else:
        dis = discount[0].text
        product_discount.append(dis)
product_discount
```

```python
products_names
```

```python
import pandas as pd
```

```python
new_data = pd.DataFrame({"Product Names": products_names, "Price ₹":product_price, "Discount %": product_discount})
```

```python
new_data.head()
```

```python
new_data["Discount %"] = new_data["Discount %"].str.replace('%', "", regex=True).str.replace('off', "", regex=True)
new_data
```

```python
new_data["Price ₹"] =new_data["Price ₹"].str.replace('₹', "", regex=True)
new_data
```
