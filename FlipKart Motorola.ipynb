```python
import requests
from bs4 import BeautifulSoup
base = 'https://www.flipkart.com/search?sid=czl&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DMOTOROLA'
result =requests.get(base)
```

```python
result.text
```

```python
content =result.text
```

```python
sp=BeautifulSoup(content, 'lxml')
```

```python
print(sp.prettify())
```

```python
contents=sp.find('div', {'id': 'container'})
```

```python
# contents.prettify()
```

```python
each = contents.findAll('div', {'class':'_4ddWXP'})
```

```python
len(each)
```

```python
image_link=[]
product_name=[]
p_price =[]
o_price =[]
category =[]
p_off =[]
rating =[]
t_rating =[]
for i in each:
    image=i.findAll('img', {'class':'_396cs4'})
    if len(image)==0:
        image_link.append("No Link Available")
    else:
        image_src_attribute=image[0].get('src')
        image_link.append(image_src_attribute)
    prod_name =i.find('a', {'class': 's1Q9rs'})
    if prod_name:
        product_name.append(prod_name.text)
    product_price=i.find('div', {'class':'_30jeq3'})
    if product_price:
        p_price.append(product_price.text)
    original_price =i.find('div', {'class': '_3I9_wc'})
    if original_price:
        o_price.append(original_price.text)
    cat=i.find('div', {'class': '_3Djpdu'})
    if cat:
        category.append(cat.text)
    percent_off =i.find('div', {'class': '_3Ay6Sb'})
    if percent_off:
        p_off.append(percent_off.text)
    rate= i.find('span', {'class': '_1lRcqv'})
    if rate:
        rating.append(rate.text)
    total_rating=i.find('span', {'class': '_2_R_DZ'})
    if total_rating:
        t_rating.append(total_rating.text)
t_rating
```

```python
import pandas as pd
n = range(1, (len(t_rating)+1))
ind=pd.Index(n)
dataset = pd.DataFrame({'Product Name': product_name,
                       'category': category,
                       'Discount Price': p_price,
                       'Original Price': o_price,
                       'Percentage Off': p_off,
                       'Rating': rating,
                       'Total Rating': t_rating,
                       "Image Link": image_link}, index=ind)
```
