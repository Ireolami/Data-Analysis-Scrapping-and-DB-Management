Web Scrapping is the process of automatically extracting data from websites. it involves using software of programming code
to access web pages, retrieve the desifred information and store it for further analysis or use. Web scrapping can 
be used for various purposes such as data collection, market research, price
comparison, and more. hOwever, its essential to respect website terms of service and regulations when engatging in web scrapping activities.

Web scrapping using Beautifulsoup, web scrapping means to collect ddata from the internet. As a beginner in data science,
you must have seen CSV files on the internet distributed by some popular website e.g. Kaggle and govt websites
the data is prepared by either collecting, writing and using standard methods or by scrapping to create a CSV file so we need two
primary packages for this task, Beautiful soup, urllib and selenium. we can easily install both of these packages using the pip command -pip 
install bs4, pip install urllib, pip install selenium.
Beautiful soup is a python library used for web scrapping purposes, it simplifies the process of parsing
HTML, and XML documents, making it easier to extract data from web pages. Beautiful Soup creates a parse
tree from the HTML source code of a web page, which you can then navigate and search to extract specific
specific elements or information. BeautifulSoup relies on the structure of HTML or XML documents on a website. it works best web the website's HTML is well-fomed 
and follows standard conventions. Websites with complex or non-standard structures may require more advanced techniques
or additional libraries.

However, Websites that hevaily rely on Javascript to load content dynamically may pose challenges for Beautiful 
Soup. Since Beautiful Soup parses static Html content, it may not easily access data loaded via Javascript.
In susch cases, you might need to combine Beautiful Soup with a tool like Selenium for web scrapping.

```python
#pip install urllib
import urllib
my_url = "https://www.flipkart.com/television-store?otracker=nmenu_sub_TVs%20and%20Appliances_0_Televisions&otracker=nmenu_sub_TVs%20%26%20Appliances_0_Television"
header = {'User-Agent':
         'Chrome/23.0.127.64 Safari/537.11',
         'Accept': 'text/html, application/xhtml, appplication/xml;q=0.9,*/*;q=0.8',#It indicates that the script can accept various forms of HTML
         'Accept.Charset': '150-8859-1,utf=8,q=0.7,*;q=0.3', #Handles ISO-8859-1 and utf-8 character 
         'Accept-Encoding': 'none',
         'Accept-Language': 'en-US, en;q=0.8', #script doesn't accept compressed content.
         'Connection': 'Keep-alive'}#to maintain a persistent connection.
uClient=urllib.request.Request(my_url, headers=header)
page=uReg(uClient).read()
```

user_Agent: THis header simulates the user agent of a web browser. it tells the web server what kind of browsser is making the 
    the request.
Accept: This header specifies the types of content that the client(script)can understand.
Accept-Charset: This header specifies the character sets that your script can handle. It indicates that your script can handle ISO-8859-1 and utf-8 character encodings.
Accept-Language: This header indicates whether your script can handle content compression.
Connection: This header specifies the type of connection to used to keep the connection alive.

```python
content=containers[0].findAll("div", {'class': '_3YgSsQ'})
len(content)
```

```python
products=[]
prices = []
rating = []
discount = []
init_price=[]
for i in contain:
    product= i.findAll('a', {'class': 's1Q9rs'})
    if len(product)==0:
        products.append(np.nan)
    else:
        prod = product[0].text
        products.append(prod)
    price_container= i.findAll('div', {'class': '_30jeq3'})
    if len(price_container)==0:
        prices.append(np.nan)
    else:
        price=price_container[0].text
        prices.append(price)
    rating_container=i.findAll('span', {'class': '_2_R_DZ'})
    if len(rating_container) ==0:
        rating.append(np.nan)
    else: 
        rate =rating_container[0].text
        rating.append(rate)
    disc=i.findAll('div', {'class': '_3Ay6Sb'})
    if len(disc)==0:
        discount.append(np.nan)
    else:
        dis = disc[0].text
        discount.append(dis)
    old_price =i.findAll("div", {'class': '_3I9_wc'})
    if len(old_price)==0:
        init_price.append(np.nan)
    else:
        olding=old_price[0].text
        init_price.append(olding)
init_price
```

```python
df = pd.DataFrame({"Product Name": products, "New Price ₹":prices,"Old Price ₹": init_price, "Discount %": discount, "Rating":rating})
df
```

```python
# prettified =container[0].prettify()
# prettified
# print(soup.prettify(container[0]))
```

```python
contents= container[0].findAll('div', {'class': 'b-list-advert__gallery__item js-advert-list-item'})
len(contents)
```

```python
contain = contents[0]
# contain
```

```python
pric = contents[0].findAll('div', {'class': 'b-advert-title-inner qa-advert-title b-advert-title-inner--div'})[0].text.strip()
price = contents[0].findAll('div', {'class': 'qa-advert-price'})[0].text.strip()
price
```

```python
car_names =[]
car_descs = []
car_price =[]
car_location =[]
car_status = []
car_drive = []
state =[]
town =[]
for i in contents:
    car_name = i.findAll('div', {'class': 'b-advert-title-inner qa-advert-title b-advert-title-inner--div'})
    if len(car_name)==0:
        car_names.append(np.nan)
    else:
        name = car_name[0].text.strip()
        car_names.append(name)
    price= i.findAll('div', {'class': 'qa-advert-price'})
    if len(price)==0:
        car_price.append(np.nan)
    else:
        pric = price[0].text.strip().replace('₦', "")
        car_price.append(pric)
    description = i.findAll('div', {'class':'b-list-advert-base__description-text'})
    if len(description)==0:
        car_descs.append(np.nan)
    else:
        des =description[0].text.strip()
        car_descs.append(des)
    location = i.findAll('span', {'class': 'b-list-advert__region__text'})
    if len(location)==0:
        car_location.append(np.nan)
    else:
        loc =location[0].text.strip()
        car_location.append(loc)
    sta = i.findAll('div', {'class': 'b-list-advert-base__item-attr'})
    if len(sta) ==0:
        car_status.append(np.nan)
    else:
        status = sta[0].text.strip()
        car_status.append(status)
    
for address in car_location:
    new= address.split()
#     de = new[1:].join(",")
    state.append(new[0])
    town.append(new[1])
# car_location
# car_price
# car_names
# car_descs
# town
# car_status
```

```python
df= pd.DataFrame({"Product Name":car_names, 'Condition':car_status, 'Description': car_descs, 'Price (₦)':car_price, "Seller's State":state, 'Town': town})
df
```
